```python
data.drop(columns=["Unnamed: 12", "Pittsfield  Massachusetts  US", 'Middlesex County', '25017'], inplace=True)
```

```python
new_cols_list=["uuid", "birth_date", "death_date", "social_security_number", "driver's_license_number", "passport_number", "salutation", "first_name",
                "middle_name", "last_name", "marital_status", "maiden_family", "skin_color", "hispanic", "gender", "address", "city", "state", "postal_code","latitude","longitude",
                "family_income"]    

old_cols = data.columns

for old_col, new_col in zip(old_cols, new_cols_list):
    data = data.rename(columns={old_col: new_col})
```

```python
data
```

```python
pattern = r'^[-]?[0-9]+.[0-9]+$'

dropped_cols = [col for col in data.columns if re.match(pattern, col)]

data.drop(columns = dropped_cols, inplace=True)
data.columns
```
